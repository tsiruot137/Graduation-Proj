```python
# # 标记几个北极圈内的重要城市
# locations = {
#     "Tromsø": (18.9553, 69.6496),
#     "Murmansk": (33.0856, 68.9733),
#     "Reykjavik": (-21.9426, 64.1355),  # 虽然低于 66.5°，但接近北极圈
#     "Barrow": (-156.7886, 71.2906)
# }

locations = [ # (lon, lat)
    (-180, 90),
    (-180, 85), (-150, 85), (-120, 85), (-90, 85), (-60, 85), (-30, 85), (0, 85), (30, 85), (60, 85), (90, 85), (120, 85), (150, 85), 
    (-180, 80), (-150, 80), (-120, 80), (0, 80), (90, 80), (120, 80), (150, 80), 
    (-180, 75), (-150, 75), (-120, 75)
]
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

# 创建一个极地投影的地图（nplaea = 北极等面积投影）
plt.figure(figsize=(8, 8))
m = Basemap(projection="nplaea", boundinglat=66.5, lon_0=0, resolution="l")

# 绘制地图边界和填充
m.drawcoastlines()  # 海岸线
m.drawcountries()  # 国家边界
m.fillcontinents(color="lightgray", lake_color="lightblue")  # 填充大陆
m.drawmapboundary(fill_color="lightblue")  # 填充海洋

# 画出北极圈（纬度 66.5°）
lat_circle = 66.5
m.drawparallels([lat_circle], labels=[1], linewidth=1.2, linestyle="dashed", color="red")

for (lon, lat) in locations:
    x, y = m(lon, lat)
    plt.plot(x, y, "ro", markersize=6)

# 显示地图
plt.title("Cities near the Arctic Circle")
plt.show()
```
